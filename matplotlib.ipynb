```python
x = np.arange(-15, 15, 0.15)
y = np.arange(-15, 15, 0.15)
x, y = np.meshgrid(x, y)

r = np.sqrt(x**2 + y**2)
z = np.sin(r)

fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z)
ax.set_title("3D Graphics")
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.set_zlabel("Z axis")

plt.show()
```
